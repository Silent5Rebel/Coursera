```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
# Всё выполнено в Python 3
```

```python
# Будем использовать логнормальное распределение, сгенерируем из него выборку размером 1000

lognorm_rv = sts.lognorm(1)
sample = lognorm_rv.rvs(1000)
```

```python
# Нарисуем гистограмму выборки и наложим поверх неё теоретическую плотность распределения

x = np.linspace(0,20,100)
pdf = lognorm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.hist(sample, normed=1, bins=50)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# Посчитаем среднее и дисперсию исходного распределения

mean, var = sts.lognorm.stats(1, moments='mv')
```

```python
# Оценим распределение выборочного среднего для выборки размером 5

clt5 = [np.mean(lognorm_rv.rvs(5)) for i in range(1000)]
```

```python
# Построим гистограмму для выборочного среднего и наложим плотность соответствующего нормального распределения 

x=np.linspace(0,20,100)
norm_rv = sts.norm(mean, (var/5)**(1/2)) 
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='log-normal n=5', alpha=0.5)
plt.hist(clt5, normed=True)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# Сделаем всё то же самое для выборки размером 10

clt10 = [np.mean(lognorm_rv.rvs(10)) for i in range(1000)]
```

```python
x=np.linspace(0,20,100)
norm_rv = sts.norm(mean, (var/10)**(1/2)) 
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='log-normal n=10', alpha=0.5)
plt.hist(clt10, normed=True)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# И теперь сделаем всё то же самое для выборки размером 50

clt50 = [np.mean(lognorm_rv.rvs(50)) for i in range(1000)]
```

```python
x=np.linspace(0,10,100)
norm_rv = sts.norm(mean, (var/50)**(1/2)) #to what should it converge?
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='log-normal n=50', alpha=0.5)
plt.hist(clt50, normed=True)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# Вывод - чем больше наша выборка, тем ближе распределение выборочного среднего 
# к соответствующему нормальному распределению.
# Также с ростом размера выборки уменьшается дисперсия выборочного среднего (и нормального)
```
